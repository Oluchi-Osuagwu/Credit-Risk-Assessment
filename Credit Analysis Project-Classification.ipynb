## Feature Engineering

```python
#Separating Features into Categorical and Numerical Variables
numerical_features = data.select_dtypes(include=['number']).columns
categorical_features = data.columns.difference(numerical_features.tolist())
```

```python
numerical_features 
```

```python
categorical_features
```

```python
#creating pandas table with numerical and categorical variables
num_features = data[numerical_features]
cat_features = data[categorical_features]
```

```python
num_features
```

```python
cat_features
```

```python
#encoding the categorical feature
label_encoder = LabelEncoder()
encoded_cat_features = label_encoder.fit_transform(cat_features)
```

```python
#accessing the original categorical values from the encoded data
original_cat_features = label_encoder.inverse_transform(encoded_cat_features)
```

```python
#displaying the original categorical values
for encoded_value, original_value in zip(encoded_cat_features, original_cat_features):
    print(f"Encoded number {encoded_value} corresponds to original value: {original_value}")
```

```python
#convert encoded values back to a pandas DataFrame
new_cat_features = pd.DataFrame(encoded_cat_features, columns=['Encoded_CreditRisk'])
new_cat_features
```

```python
#concatenate the new categorical DataFrame with numerical DataFrame
data_encoded = pd.concat([num_features, new_cat_features], axis=1)

#display the updated DataFrame
data_encoded.head()
```

```python
#correlation matrix
corr_matrix = data_encoded.corr()
corr_matrix
```

```python
#correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap="RdYlBu")
plt.title("Correlation Heatmap")
plt.show()
```

```python
#dropping application ID, not necessary for the project
data_new = data_encoded.drop('Applicant_ID', axis=1)
data_new.head()
```

**Separating target from input variables and scaling the input variables**

```python
#separating the target variable from the input variable
Y = data_new['Encoded_CreditRisk']
X = data_new.drop(columns = 'Encoded_CreditRisk')
```

```python
#scaling the input variables
from sklearn.preprocessing import StandardScaler

#creating the instance of StandardScaler
ss = StandardScaler()

#scaling the input
X_scaled = ss.fit_transform(X)

#creating a new dataframe with the scaled feature
X_scaled = pd.DataFrame(X_scaled, columns = X.columns)
```

**Feature importance**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_scaled, Y, test_size = 0.1, stratify = Y, random_state = 42)

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
#fit the rf based on train data
rf.fit(x_train, y_train)
#predict the target based on test data
y_pred=rf.predict(x_test)

feature_imp = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=False)

ax = feature_imp.plot(kind='bar', figsize=(16, 6))
ax.set(ylabel='Relative Importance');
ax.set(xlabel='Feature');
```

The credit score is a strong player in credit risk analysis, I would be dropping the column

```python
New_X_scaled = X_scaled.drop(columns = 'Credit_Score')
New_X_scaled
```

```python
#splitting again because of the drop
x_train, x_test, y_train, y_test = train_test_split(New_X_scaled, Y, test_size = 0.1, stratify = Y, random_state = 42)
```

## Model Development

### Random Forest

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, ConfusionMatrixDisplay, precision_score, recall_score
```

```python
rf = RandomForestClassifier()
#Fit the rf based on train data
rf.fit(x_train, y_train)
#Predict the target based on test data
y_pred=rf.predict(x_test)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)

print ('Report : ')
print (classification_report(y_test, y_pred))

```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)

plt.show()
```

**Cross validation**

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(rf, x_train, y_train, scoring="f1_micro", cv=10)

df_rf_F1 = pd.DataFrame(scores, columns = ['F1-Score'])
df_rf_F1.plot(kind = 'bar')
plt.title('F1 Scores obtained over 10-Fold Cross-validation \n using RF Classification with an optimal set of parameters', fontsize = 16)
```

### KNN ( K-Nearest Neighbors)

```python
from sklearn import neighbors
from sklearn.neighbors import KNeighborsClassifier
```

```python
#storing the accuracy values of different k
accuracy = []
#storing the values of k
K = []
#storing precison values of different k
precision = []
#storing recall values of different k
recall = []
#For loop for eack k
for k in range (10, 31, 10):
    #Define knn as KNeighborsClassifier for each K intiated from for loop
    knn = KNeighborsClassifier(n_neighbors = k)
    #Fit the knn based on train data
    knn.fit(x_train, y_train)
    #predict the target based on test data
    y_pred = knn.predict(x_test)
    #store the value of k
    K.append(k)
    #store the value of accuracy
    accuracy.append(accuracy_score(y_test, y_pred))
    #store the value of precision of each target
    precision.append(precision_score(y_test, y_pred, average = None))
    #store the value ofrecall each target
    recall.append(recall_score(y_test, y_pred, average = None))
```

```python
# Plot precision for each target
for i, target in enumerate(Y.unique()):
    plt.plot(K, [p[i] for p in precision], label=target)

plt.xlabel('K')
plt.ylabel('Precision')
plt.title('Precision for Each Target')
plt.legend()
plt.show()
```

```python
# Plot recall for each target
for i, target in enumerate(Y.unique()):
    plt.plot(K, [r[i] for r in recall], label=target)

plt.xlabel('K')
plt.ylabel('Recall')
plt.title('Recall for Each Target')
plt.legend()
plt.show()
```

```python
# Plot accuracy for each target
for i, target in enumerate(Y.unique()):
    plt.plot(K, accuracy)
    
plt.xlabel('K')
plt.ylabel('Accuracy')
plt.title('Accuracy for Target')
plt.show()
```

From the accuracy, recall and precision, k = 10 is the best k

**Using Optimal K for the Classification**

```python
#Use knn at K=10
knn = KNeighborsClassifier(n_neighbors=10)
    
#Fit the knn based on train data
knn.fit(x_train, y_train) 
    
#Predict the target based on test data
y_pred = knn.predict(x_test)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)

print ('Report : ')
print (classification_report(y_test, y_pred))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)

plt.show()
```

**Performing Cross Validation**

```python
scores = cross_val_score(knn, x_train, y_train, scoring="f1_micro", cv=10)
```

```python
df_knn_F1 = pd.DataFrame(scores, columns = ['F1-Score'])
df_knn_F1.plot(kind = 'bar')
plt.title('F1 Scores obtained over 10-Fold Cross-validation \n using KNN Classification with an optimal set of parameters', fontsize = 16)
```
