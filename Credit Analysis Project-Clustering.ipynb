### Unsupervised Machine Learning

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt, numpy as np
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import AgglomerativeClustering
from matplotlib.colors import ListedColormap
from sklearn import metrics
import warnings
import sys
if not sys.warnoptions:
    warnings.simplefilter("ignore")
np.random.seed(42)
```

```python
df.head()
```

```python
numerical_features = data.select_dtypes(include=['number']).columns
categorical_features = data.columns.difference(numerical_features.tolist())
```

```python
numerical_features = df[numerical_features]
categorical_features = df[categorical_features]
```

```python
from sklearn.preprocessing import StandardScaler

ss =StandardScaler()
    
for col in numerical_features.columns:
    numerical_features[col] = ss.fit_transform(numerical_features[[col]]).squeeze()
```

```python
#Enconding the CategoricalFeatures
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
categorical_features = categorical_features.astype(str).apply(LabelEncoder().fit_transform)
```

```python
df = pd.concat([numerical_features, categorical_features], axis = 1)
```

```python
df.head()
```

```python
corr_mat = df.corr()

# Strip the diagonal for future examination
for x in range(corr_mat.shape[0]):
    corr_mat.iloc[x,x] = 0.0
    
corr_mat
```

```python
corr_mat.abs().idxmax()
```

```python
#Determining number of cluster using Elbow Method
from sklearn.cluster import KMeans
inertia = []
list_num_clusters = list(range(1,11))
for num_clusters in list_num_clusters:
    km = KMeans(n_clusters=num_clusters)
    km.fit(df)
    inertia.append(km.inertia_)
plt.plot(list_num_clusters,inertia)
plt.scatter(list_num_clusters,inertia)
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia');
### END SOLUTION
```

```python
num_clusters = 4
km = KMeans(n_clusters=num_clusters,random_state=10,n_init=1) # n_init, number of times the K-mean algorithm will run
yhat_km = km.fit_predict(df)
data["CLUSTERS"] = yhat_km
```

```python
data.head()
```

```python
sns.scatterplot(x='Debt_Income_Ratio', y='Loan_Amt', hue='CLUSTERS', data=df, palette='viridis')

# Customize the plot
plt.xlabel('Credit Risk')
plt.ylabel('Loan Amount')
plt.title('K-means Clustering')

# Show the plot
plt.show()
```

```python
pal = ["#682F2F","#B9C0C9", "#9F8A78","#F3AB60"]
pl = sns.countplot(x=data["CLUSTERS"], palette= pal)
pl.set_title("Distribution Of The Clusters")
plt.show()
```

```python
fig = plt.figure(figsize=(10,8))
pl = sns.countplot(x=data["Credit_Risk"],hue=data["CLUSTERS"], palette= pal)
pl.set_title("CLUSTERS BY Credit_Risk")
pl.set_xlabel("TENURE")
plt.show()
```

```python
#Initiating the Agglomerative Clustering model 
AC = AgglomerativeClustering(n_clusters=4)
# fit model and predict clusters
yhat_AC = AC.fit_predict(df)
data["AC_Clusters"] = yhat_AC
#Adding the Clusters feature to the orignal dataframe.
```

### Applying PCA

```python
pca = PCA(n_components=3)
pca.fit(df)
PCA_df = pd.DataFrame(pca.transform(df), columns=(["col1","col2", "col3"]))
PCA_df.describe().T
```

```python
num_clusters = 4
km = KMeans(n_clusters=num_clusters,random_state=10,n_init=1) # n_init, number of times the K-mean algorithm will run
yhat_km = km.fit_predict(PCA_df)
PCA_df['Clusters'] = yhat_km
data["ClustersPCA"] = yhat_km
```
